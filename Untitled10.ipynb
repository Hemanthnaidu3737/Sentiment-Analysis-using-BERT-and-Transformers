```python
import gradio as gr
from transformers import pipeline
import torch
import time
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Enable GPU if available
device = 0 if torch.cuda.is_available() else -1
```

```python
# Load pre-trained sentiment analysis model
sentiment_pipeline = pipeline("text-classification", model="nlptown/bert-base-multilingual-uncased-sentiment", device=device)
```

```python
# Define intent-based responses
intent_responses = {
    "positive": ["That's great to hear! 😊", "Awesome! Keep spreading positivity! 🎉"],
    "neutral": ["Okay, got it! 😐", "Thanks for sharing!"],
    "negative": ["I'm sorry to hear that. 😔", "Let's try to make things better! 🌟"]
}
```

```python
def predict_sentiment(user_input):
    """Predict sentiment and return response with a progress bar."""
    if isinstance(user_input, list):  # Batch Processing for Multiple Inputs
        results = [sentiment_pipeline(text)[0] for text in user_input]
        sentiments = []

        for res in results:
            label = res["label"]
            if "5 stars" in label or "4 stars" in label:
                sentiments.append("positive")
            elif "3 stars" in label:
                sentiments.append("neutral")
            else:
                sentiments.append("negative")

        # Generate bar chart
        labels, counts = np.unique(sentiments, return_counts=True)
        fig, ax = plt.subplots()
        ax.bar(labels, counts, color=['green', 'yellow', 'red'])
        ax.set_title("Sentiment Distribution")
        ax.set_ylabel("Count")

        return "Batch analysis complete!", fig

    else:  # Single Input Processing
        time.sleep(1.5)  # Simulate loading effect
        result = sentiment_pipeline(user_input)
        label = result[0]["label"]
        score = result[0]["score"]

        # Map model labels to sentiments
        if "5 stars" in label or "4 stars" in label:
            sentiment = "positive"
        elif "3 stars" in label:
            sentiment = "neutral"
        else:
            sentiment = "negative"

        response = intent_responses[sentiment][0]

        return f"Sentiment: {sentiment.capitalize()}\nConfidence: {score:.2f}\nResponse: {response}"
```

```python
# Function to process file input
def process_file(file):
    if file:
        with open(file.name, 'r', encoding='utf-8') as f:
            texts = f.readlines()
        return texts
    return []

```

```python
import gradio as gr

# Create Gradio interface
interface = gr.Interface(
    fn=predict_sentiment,
    inputs=gr.Textbox(lines=3, placeholder="Enter your text here..."),
    outputs=gr.Textbox(),
    title="Sentiment Analysis Using BERT And Transformers",
    description="Enter a sentence, and the chatbot will analyze its sentiment and generate a response."
)

# Launch the UI
interface.launch(share=True)
```
